```python
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
from matplotlib import pylab
import datetime as dt
import pandas as pd
import calendar
%matplotlib inline

csv_path13_1 = "dataset\\일자별 위탁판매현황(2013년1월).csv"
csv_path13_2 = "dataset\\일자별 위탁판매현황(2013년2월).csv"
csv_path13_3 = "dataset\\일자별 위탁판매현황(2013년3월).csv"
csv_path13_4 = "dataset\\일자별 위탁판매현황(2013년4월).csv"
csv_path13_5 = "dataset\\일자별 위탁판매현황(2013년5월).csv"
csv_path13_6 = "dataset\\일자별 위탁판매현황(2013년6월).csv"
csv_path13_7 = "dataset\\일자별 위탁판매현황(2013년7월).csv"
csv_path13_8 = "dataset\\일자별 위탁판매현황(2013년8월).csv"
csv_path13_9 = "dataset\\일자별 위탁판매현황(2013년9월).csv"
csv_path13_10= "dataset\\일자별 위탁판매현황(2013년10월).csv"
csv_path13_11= "dataset\\일자별 위탁판매현황(2013년11월).csv"
csv_path13_12= "dataset\\일자별 위탁판매현황(2013년12월).csv"


#csv 읽어옴
read13_1 = pd.read_csv(csv_path13_1, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_2 = pd.read_csv(csv_path13_2, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_3 = pd.read_csv(csv_path13_3, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_4 = pd.read_csv(csv_path13_4, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_5 = pd.read_csv(csv_path13_5, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_6 = pd.read_csv(csv_path13_6, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_7 = pd.read_csv(csv_path13_7, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_8 = pd.read_csv(csv_path13_8, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_9 = pd.read_csv(csv_path13_9, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_10 = pd.read_csv(csv_path13_10, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_11 = pd.read_csv(csv_path13_11, engine='python',error_bad_lines=False, warn_bad_lines=False)
read13_12 = pd.read_csv(csv_path13_12, engine='python',error_bad_lines=False, warn_bad_lines=False)

#데이터프레임 생성
df13_1 = DataFrame(read13_1)
df13_2 = DataFrame(read13_2)
df13_3 = DataFrame(read13_3)
df13_4 = DataFrame(read13_4)
df13_5 = DataFrame(read13_5)
df13_6 = DataFrame(read13_6)
df13_7 = DataFrame(read13_7)
df13_8 = DataFrame(read13_8)
df13_9 = DataFrame(read13_9)
df13_10= DataFrame(read13_10)
df13_11= DataFrame(read13_11)
df13_12= DataFrame(read13_12)

#데이터프레임 연결한다.
temp = pd.concat([df13_1,df13_2]) 
temp = pd.concat([temp, df13_3])
temp = pd.concat([temp, df13_4])
temp = pd.concat([temp, df13_5])
temp = pd.concat([temp, df13_6])
temp = pd.concat([temp, df13_7])
temp = pd.concat([temp, df13_8])
temp = pd.concat([temp, df13_9])
temp = pd.concat([temp, df13_10])
temp = pd.concat([temp, df13_11])
df = pd.concat([temp, df13_12])

#어종상태명이 '(활)' 인 row의 위판일, 코드명, 상태명, 단가 열을 가져온다.
df = df.loc[df.어종상태명=="(활)",["위판일자","수산물표준코드명","어종상태명","위판단가"]]
#코드명이 '가자미류'인 row만 선택한다.
df = df.loc[df.수산물표준코드명=="가자미류",["위판일자","수산물표준코드명","어종상태명","위판단가"]]
#그룹별 평균을 구한다. 여기서는 위판일자 별 위판단가 평균
df_mean = df.groupby(['위판일자','수산물표준코드명'], as_index=False).mean() 

#그래프 월별 색깔 지정 (10가지 색상)
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

#한글 폰트 지정
plt.rcParams["font.family"] = 'NanumBarunGothic'
plt.rcParams["font.size"] = 10

#x,y좌표 라벨, 표 크기 지정
plt.figure(figsize=(24,10))
plt.xlabel('날짜')
plt.ylabel('가격(원)')

#[plot data] x좌표 : 날짜, y좌표 : 일 평균 가격
x,y = [],[]
for line in df_mean.위판일자:
    times = dt.datetime.strptime(str(line),'%Y%m%d')
    x.append(times)
for line in df_mean.위판단가:
    y.append(line)

#강조 구간
for i in range(1,13):
    span_start = dt.datetime(2013, i, 1)
    end_day = calendar.monthrange(2013, i)[1] # i월달의 마지막 일을 가져온다.
    span_end = dt.datetime(2013, i, end_day)
    plt.axvspan(span_start, span_end, facecolor=colors[i%10], alpha=0.5)

plt.title("가자미류 13년도 일 평균 가격")
plt.plot(x,y,linestyle='-',c="m" ,label="value")
plt.show()




```
